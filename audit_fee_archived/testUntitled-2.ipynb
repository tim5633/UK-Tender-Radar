```python
import requests
import base64
import pandas as pd
import os
import fitz
import cv2
import pytesseract
import re
from pytesseract import Output
from difflib import SequenceMatcher

# -----------------------------
# Config
# -----------------------------
API_KEY = "d6c35590-9518-4569-b335-cbc90111d5dd"
auth_str = base64.b64encode(f"{API_KEY}:".encode()).decode()
headers = {"Authorization": f"Basic {auth_str}"}

SEARCH_QUERY = "British Airways"
OUTPUT_DIR = "/Users/timliu/Desktop/uk_accounts_pdfs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

TMP_DIR = "/tmp/audit_debug"
os.makedirs(TMP_DIR, exist_ok=True)

AUDIT_KEYWORDS = [
    "audit", "auditor", "auditors",
    "statutory audit", "statutory auditor",
    "fees to auditors", "auditors remuneration",
    "remuneration of auditors", "audit services",
    "audit and assurance", "external auditor",
    "independent auditor", "audit fees"
]

# -----------------------------
# Helper Functions (Companies House)
# -----------------------------
def search_companies(query, start_index=0):
    url = f"https://api.company-information.service.gov.uk/search/companies?q={query}&start_index={start_index}"
    r = requests.get(url, headers=headers)
    if r.status_code != 200:
        print("Error:", r.text)
        return [], 0
    data = r.json()
    companies = []
    total_results = data.get("total_results", 0)
    if "items" in data:
        companies = [item for item in data["items"] if item["company_status"] == "active"]
    return companies, total_results

def get_filing_history(company_number):
    url = f"https://api.company-information.service.gov.uk/company/{company_number}/filing-history"
    r = requests.get(url, headers=headers)
    if r.status_code != 200:
        return []
    data = r.json()
    return data.get("items", [])

def filter_accounts(filings):
    return [f for f in filings if f.get("category") == "accounts"]

def download_pdf(filing, company_number):
    doc_meta_url = filing["links"].get("document_metadata")
    if not doc_meta_url:
        return None
    meta = requests.get(doc_meta_url, headers=headers).json()
    pdf_url = meta.get("links", {}).get("document")
    if not pdf_url:
        return None

    r = requests.get(pdf_url, headers=headers)
    pdf_path = os.path.join(OUTPUT_DIR, f"{company_number}_{filing['date']}.pdf")
    with open(pdf_path, "wb") as f:
        f.write(r.content)
    return pdf_path

# -----------------------------
# Helper Functions (OCR / Audit Extraction)
# -----------------------------
def fuzzy_match(a, b, threshold=0.7):
    return SequenceMatcher(None, a, b).ratio() >= threshold

def pdf_to_images(pdf_path, zoom=3):
    doc = fitz.open(pdf_path)
    image_paths = []
    for page in doc:
        mat = fitz.Matrix(zoom, zoom)
        pix = page.get_pixmap(matrix=mat)
        img_path = os.path.join(TMP_DIR, f"{os.path.basename(pdf_path)}_p{page.number}.png")
        pix.save(img_path)
        image_paths.append(img_path)
    return image_paths

def preprocess(img_path):
    img = cv2.imread(img_path, 0)
    img = cv2.threshold(img, 150, 255, cv2.THRESH_BINARY)[1]
    img = cv2.medianBlur(img, 3)
    return img

def find_audit_blocks(img):
    data = pytesseract.image_to_data(img, output_type=Output.DICT)
    boxes = []
    for i, txt in enumerate(data["text"]):
        t = txt.lower().strip()
        if not t:
            continue
        for kw in AUDIT_KEYWORDS:
            if kw in t or fuzzy_match(t, kw):
                x, y, w, h = data["left"][i], data["top"][i], data["width"][i], data["height"][i]
                boxes.append((x, y, w, h))
                break
    return boxes

def crop_region(img, box, pad=120):
    x, y, w, h = box
    H, W = img.shape
    x1 = max(0, x - pad)
    y1 = max(0, y - pad)
    x2 = min(W, x + w + pad*3)
    y2 = min(H, y + h + pad*2)
    return img[y1:y2, x1:x2]

def extract_info(text):
    text = text.lower()
    auditor, fee, unit = "", "", ""
    m = re.search(r"(auditor|auditors|statutory auditor|external auditor)[:\s\-]*([a-z &.,]{3,})", text)
    if m:
        auditor = m.group(2).strip()
    m = re.search(r"(audit|auditors).*?(fee|fees|remuneration).*?([\£$]?\s?[\d,]+)", text)
    if m:
        num = re.search(r"[\d,]+", m.group(0))
        if num:
            fee = num.group()
    if re.search(r"£\s?'?000|thousand", text):
        unit = "thousand_gbp"
    elif "£" in text or "gbp" in text:
        unit = "gbp"
    return auditor, fee, unit

def process_pdf(pdf_path):
    results = []
    images = pdf_to_images(pdf_path)
    for img_path in images:
        img = preprocess(img_path)
        boxes = find_audit_blocks(img)
        for i, box in enumerate(boxes):
            crop = crop_region(img, box)
            crop_path = os.path.join(TMP_DIR, f"crop_{os.path.basename(img_path)}_{i}.png")
            cv2.imwrite(crop_path, crop)
            text = pytesseract.image_to_string(crop)
            auditor, fee, unit = extract_info(text)
            if fee or auditor:
                results.append({
                    "pdf": pdf_path,
                    "auditor": auditor,
                    "audit_fee": fee,
                    "unit": unit,
                    "text": text,
                    "crop_path": crop_path
                })
    return results

# -----------------------------
# Main Loop
# -----------------------------
all_results = []

companies, total = search_companies(SEARCH_QUERY)
print(f"總共 {total} 公司，先抓 {len(companies)} 公司作測試")

for company in companies:
    company_number = company["company_number"]
    company_name = company["title"]

    filings = get_filing_history(company_number)
    accounts_filings = filter_accounts(filings)

    for filing in accounts_filings:
        pdf_path = download_pdf(filing, company_number)
        if pdf_path:
            print(f"{company_name} PDF downloaded: {pdf_path}")
            audit_info = process_pdf(pdf_path)
            for row in audit_info:
                row["company_number"] = company_number
                row["company_name"] = company_name
                row["report_date"] = filing["date"]
                all_results.append(row)

# -----------------------------
# Save CSV
# -----------------------------
df = pd.DataFrame(all_results)
# csv_path = os.path.join(OUTPUT_DIR, "uk_audit_extracted.csv")
# df.to_csv(csv_path, index=False)
print(f"Done! CSV saved to {csv_path}")
```

```python
import requests
import base64
import pandas as pd
import os
import fitz  # PyMuPDF
import re

# -----------------------------
# Config
# -----------------------------
API_KEY = "d6c35590-9518-4569-b335-cbc90111d5dd"
auth_str = base64.b64encode(f"{API_KEY}:".encode()).decode()
headers = {"Authorization": f"Basic {auth_str}"}

SEARCH_QUERY = "British Airways"
OUTPUT_DIR = "/Users/timliu/Desktop/uk_accounts_pdfs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# -----------------------------
# Helper Functions (Companies House)
# -----------------------------
def search_companies(query, start_index=0):
    url = f"https://api.company-information.service.gov.uk/search/companies?q={query}&start_index={start_index}"
    r = requests.get(url, headers=headers)
    if r.status_code != 200:
        print("Error:", r.text)
        return [], 0
    data = r.json()
    companies = [item for item in data.get("items", []) if item["company_status"] == "active"]
    total_results = data.get("total_results", 0)
    return companies, total_results

def get_filing_history(company_number):
    url = f"https://api.company-information.service.gov.uk/company/{company_number}/filing-history"
    r = requests.get(url, headers=headers)
    if r.status_code != 200:
        return []
    data = r.json()
    return data.get("items", [])

def filter_accounts(filings):
    return [f for f in filings if f.get("category") == "accounts"]

# -----------------------------
# PDF Text Extraction
# -----------------------------
def extract_text_from_pdf(pdf_path):
    try:
        doc = fitz.open(pdf_path)
        full_text = ""
        for page in doc:
            full_text += page.get_text()
        return full_text.lower()
    except Exception as e:
        print("PDF extraction error:", e)
        return ""

# -----------------------------
# Extract Audit Info
# -----------------------------
def extract_audit_info(filing, pdf_text=None):
    text_fields = []

    # 先用 description_values
    if "description_values" in filing:
        for k, v in filing["description_values"].items():
            text_fields.append(str(v).lower())

    # 如果有 PDF text，也加入
    if pdf_text:
        text_fields.append(pdf_text.lower())

    combined_text = " ".join(text_fields)

    auditor, fee, unit = "", "", ""

    # 抓審計師名字
    m = re.search(r"(auditor|auditors|statutory auditor|external auditor)[:\s\-]*([a-z &.,]{3,})", combined_text)
    if m:
        auditor = m.group(2).strip()

    # 抓審計費
    m = re.search(r"(audit|auditors).*?(fee|fees|remuneration).*?([\£$]?\s?[\d,]+)", combined_text)
    if m:
        num = re.search(r"[\d,]+", m.group(0))
        if num:
            fee = num.group()
    
    if re.search(r"£\s?'?000|thousand", combined_text):
        unit = "thousand_gbp"
    elif "£" in combined_text or "gbp" in combined_text:
        unit = "gbp"

    return auditor, fee, unit

# -----------------------------
# Download PDF (Optional)
# -----------------------------
def download_pdf(filing, company_number):
    doc_meta_url = filing["links"].get("document_metadata")
    if not doc_meta_url:
        return None
    meta = requests.get(doc_meta_url, headers=headers).json()
    pdf_url = meta.get("links", {}).get("document")
    if not pdf_url:
        return None

    pdf_path = os.path.join(OUTPUT_DIR, f"{company_number}_{filing['date']}.pdf")
    r = requests.get(pdf_url, headers=headers)
    with open(pdf_path, "wb") as f:
        f.write(r.content)
    return pdf_path

# -----------------------------
# Main Loop
# -----------------------------
all_results = []

companies, total = search_companies(SEARCH_QUERY)
print(f"總共 {total} 公司，先抓 {len(companies)} 公司作測試")

for company in companies:
    company_number = company["company_number"]
    company_name = company["title"]

    filings = get_filing_history(company_number)
    accounts_filings = filter_accounts(filings)

    for filing in accounts_filings:
        # 可選下載 PDF，如果想用 PDF 文字抓
        pdf_path = download_pdf(filing, company_number)
        pdf_text = extract_text_from_pdf(pdf_path) if pdf_path else None

        auditor, fee, unit = extract_audit_info(filing, pdf_text)

        if auditor or fee:
            all_results.append({
                "company_number": company_number,
                "company_name": company_name,
                "report_date": filing["date"],
                "auditor": auditor,
                "audit_fee": fee,
                "unit": unit,
                "pdf_path": pdf_path
            })

# -----------------------------
# Save CSV
# -----------------------------
df = pd.DataFrame(all_results)
# csv_path = os.path.join(OUTPUT_DIR, "uk_audit_extracted.csv")
# df.to_csv(csv_path, index=False)
print(f"Done! CSV saved to {csv_path}")
```
